### Обработка пропусков

Резюме:

- у нас 23% пропусков в переменной 'balance' (баланс счета), это большой объем данных;
- есть связь пропусков баланса счета с другими показателями:  чем ниже балл собственности и чем меньше продуктов использует клиент, тем выше вероятность пропущенных данных. С переменной 'products' связь скорее слабая(коэффициент корреляции 0,48), а вот с количеством баллов собственности - средняя или сильная (0,71);
- Медианное значение 'balance' заметно различается в зависимости от количества баллов собственности: 

Пропуски в поле баланс счета относятся к неслучайным (MNAR). 
Так как есть корреляция пропусков баланса счета с количеством баллов собственности, а также средний баланс счета заметно различается в зависимости от баллов собственности, 
<i>заполним пропуски в поле 'balance' медианным значением для групп по количеству баллов собственности. </i>


Пропущенные данные в столбце "возраст" не скореллированы с другими признаками, таковых менее 1%, они, скорее всего, возникли из-за технического сбоя и относятся к категории совершенно случайных пропусков (MCAR). 
Пропуски возраста заполним медианой для групп по количеству баллов собственности(аналогично пропускам баланса счета).

```python
for b in banki['equity'].unique():
    banki.loc[(banki['equity'] == b) & (banki['balance'].isna()), 'balance'] = \
    banki.loc[(banki['equity'] == b), 'balance'].median()
```

```python
for b in banki['equity'].unique():
    banki.loc[(banki['equity'] == b) & (banki['age'].isna()), 'age'] = \
    banki.loc[(banki['equity'] == b), 'age'].median()
```

## Выбор гипотез и тестов - 2
На втором шаге проверим гипотезы о <u><i>равенстве оттока среди мужчин и женщин</i></u>; <u><i>среди пользователей 1-2 и в группе кто использует 3 и более банковских продукта</i></u>. 
Для этого мы используем тест на сравнение пропорций в двух генеральных совокупностях.

## Проверка гипотезы о равенстве  среднего баланса счета у отточных и лояльных клиентов 

Гипотеза 1:

    H0: средний баланс счета в группе отточных и лояльных клиентов одинаков
    H1: средний баланс счета в группе отточных и лояльных клиентов различается

```python
sample1_churn = banki_filt.query('churn == 1')['balance']
sample2_notchurn = banki_filt.query('churn == 0')['balance']

alpha = 0.05  # критический уровень статистической значимости

results = st.mannwhitneyu(sample1_churn, sample2_notchurn)

print('p-значение: ', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу, средний баланс счета в группе отточных и лояльных клиентов различается')
else:
    print(
        'Нет оснований отвергнуть нулевую гипотезу, нет значимой разницы среднего баланса счета для отточных и лояльных клиентов'
    ) 
    
```

## Проверка гипотезы о равенстве среднего дохода  у отточных и лояльных клиентов 

Гипотеза 2: 

    H0: средний доход в группе отточных и лояльных клиентов одинаков
    H1: средний доход в группе отточных и лояльных клиентов значимо различается

```python
sample3_churn = banki_filt.query('churn == 1')['est_salary']
sample4_notchurn = banki_filt.query('churn == 0')['est_salary']

alpha = 0.05  # критический уровень статистической значимости

results = st.mannwhitneyu(sample3_churn, sample4_notchurn)

print('p-значение: ', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу, средний доход в группе отточных и лояльных клиентов различается')
else:
    print(
        'Нет оснований отвергнуть нулевую гипотезу, нет значимой разницы среднего дохода для отточных и лояльных клиентов'
    ) 
    
```

Нет статистически значимой разницы в доходах лояльных и отточных клиентов. При этом, тест подтверждает, что у отточных клиентов баланс счета в среднем выше, чем у лояльных. 

## Проверка гипотезы о равенстве оттока среди мужчин и женщин

    Гипотеза 3: 
    H0: Отток среди мужчин равен оттоку среди женщин.
    H1: Отток среди мужчин отличается от оттока среди женщин.

```python
#Посчитаем необходимые значения для сравнения:
n_m = len(banki_filt.query('gender_dummy == 1')) #Количество мужчин в выборке
n_f = len (banki_filt.query('gender_dummy == 0')) # Количество женщин в выборке
churn_m =banki_filt.query('gender_dummy ==1')['churn'].sum() # Количество мужчин, которые ушли
churn_f = banki_filt.query('gender_dummy ==0')['churn'].sum() # Количество женщин, которые ушли
```

```python
successes = np.array([churn_m, churn_f]) # Запишем количество мужчин и женщин, отказавшихся от услуг банка
trials = np.array([n_m, n_f]) # Запишем количество мужчин и женщин в выборке

alpha = .05 # критический уровень статистической значимости

# пропорция успехов в первой группе:
p1 = successes[0]/trials[0]

# пропорция успехов во второй группе:
p2 = successes[1]/trials[1]

# пропорция успехов в комбинированном датасете:
p_combined = (successes[0] + successes[1]) / (trials[0] + trials[1])

# разница пропорций в датасетах
difference = p1 - p2 

# считаем статистику в ст.отклонениях стандартного нормального распределения
z_value = difference / mth.sqrt(p_combined * (1 - p_combined) * (1/trials[0] + 1/trials[1]))

# задаем стандартное нормальное распределение (среднее 0, ст.отклонение 1)
distr = st.norm(0, 1)  

p_value = (1 - distr.cdf(abs(z_value))) * 2

print('p-значение: ', p_value)

if p_value < alpha:
    print('Отвергаем нулевую гипотезу: доли отточных клиентов среди мужчин и женщин значимо различаются')
else:
    print(
        'Не получилось отвергнуть нулевую гипотезу, нет оснований считать доли отточных клиентов среди мужчин и женщин разными'
    ) 
```

## Проверка гипотезы о равенстве оттока среди клиентов с разным количеством банковских продуктов

Гипотеза 4:

    H0: Отток среди пользователей 3-5  банковских продуктов равен оттоку среди пользователей 1-2 продуктов.
    H1: Отток среди пользователей 3-5  банковских продуктов выше оттока среди пользователей 1-2 продуктов.


```python
#Посчитаем необходимые значения для сравнения:
n_1 = len(banki_filt.query('products >= 3'))  #количество наблюдений в группе с 3-5 продуктами
n_2 = len(banki_filt.query('products <3'))  #количество наблюдений в группе с 1-2 продуктами
churn_1 =banki_filt.query('products >= 3')['churn'].sum()  # отток в  группе с 3-5 продуктами
churn_2 = banki_filt.query('products <3')['churn'].sum() # отток в группе с 1-2 продуктами
```

```python
successes = np.array([churn_1, churn_2]) # Запишем количество  отказавшихся от услуг банка в двух группах
trials = np.array([n_1, n_2]) # Запишем количество наблюдений в двух группах

alpha = .05 # критический уровень статистической значимости

# пропорция успехов в первой группе:
p1 = successes[0]/trials[0]

# пропорция успехов во второй группе:
p2 = successes[1]/trials[1]

# пропорция успехов в комбинированном датасете:
p_combined = (successes[0] + successes[1]) / (trials[0] + trials[1])

# разница пропорций в датасетах
difference = p1 - p2 

# считаем статистику в ст.отклонениях стандартного нормального распределения
z_value = difference / mth.sqrt(p_combined * (1 - p_combined) * (1/trials[0] + 1/trials[1]))

# задаем стандартное нормальное распределение (среднее 0, ст.отклонение 1)
distr = st.norm(0, 1)  

p_value = (1 - distr.cdf(abs(z_value))) * 2

print('p-значение: ', p_value)

if p_value < alpha:
    print('Отвергаем нулевую гипотезу: есть значимая разница оттока для клиентов, использующих 1-2 и 3 и более банковских продуктов')
else:
    print(
        'Не получилось отвергнуть нулевую гипотезу, нет оснований считать доли оттока в кгруппах с различным количеством продуктов разными'
    ) 
```

## Проверка гипотезы о равенстве среднего возраста отточных и лояльных клиентов среди клиентов допенсионного возраста (до 63 лет)

Гипотеза 5: 
 


    H0: средний возраст отточных и лояльных клиентов допенсионного возраста(до 63 лет) одинаков
    H1: средний возраст в группе отточных клиентов выше, чем в группе лояльных клиентов допенсионного возраста(до 63 лет)

```python
#сформируем выборки для проверки
sample1_churn = banki_filt.query('churn == 1 & age <= 63')['age']
sample2_notchurn = banki_filt.query('churn == 0 & age <=63')['age']

alpha = 0.05

# проводим ttest для двух совокупностей, в качестве альтернативной используем одностороннюю гипотезу «больше»
results = st.ttest_ind(sample1_churn, sample2_notchurn, alternative='greater', equal_var = False)

print(f'p-value: {results.pvalue}')

# проверяем p-value
if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу, средний возраст отточных клиентов выше, чем  у лояльных в группе до 63 лет')
else:
    print('Нет оснований отвергнуть нулевую гипотезу, средний возраст отточных клиентов равен возрасту лояльных в группе до 63 лет')
```
